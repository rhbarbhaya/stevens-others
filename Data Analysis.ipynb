```python
# Importing necessary files and libraries
import pandas as pd
pd.options.display.max_columns = None
import numpy as np

raw_dataset = pd.read_csv("loan_interest_rates.csv", low_memory = False)
raw_dataset.head()
```

```python
# Initial cleaning, to keep the important columns for regression analysis
raw_dataset["X8"] = raw_dataset["X8"].replace(np.nan, "NULL")
raw_dataset["X9"] = raw_dataset["X9"].replace(np.nan, "NULL")
raw_dataset["X10"] = raw_dataset["X10"].replace(np.nan, "NULL")
raw_dataset["X12"] = raw_dataset["X12"].replace(np.nan, "NULL")
raw_dataset["X13"] = raw_dataset["X13"].replace(np.nan, "NULL")
raw_dataset["X16"] = raw_dataset["X16"].replace(np.nan, "NULL")
raw_dataset["X18"] = raw_dataset["X18"].replace(np.nan, "NULL")
raw_dataset["X23"] = raw_dataset["X23"].replace(np.nan, 0)
raw_dataset["X24"] = raw_dataset["X24"].replace(np.nan, 0) 
raw_dataset.head()
```

```python
# Dropping some incomplete columns for data analysis and regession
cleaner_dataset = raw_dataset.dropna()
cleaner_dataset.shape
```

```python
#checking unique values. Will need to convert some non-numerical columns to numerical for regression
print("Column X7", cleaner_dataset["X7"].unique(), "--> Length -->", len(cleaner_dataset["X7"].unique()))
print("Column X8", cleaner_dataset["X8"].unique(), "--> Length -->", len(cleaner_dataset["X8"].unique()))
print("Column X9", cleaner_dataset["X9"].unique(), "--> Length -->", len(cleaner_dataset["X9"].unique()))
print("Column X19", cleaner_dataset["X19"].unique(), "--> Length -->", len(cleaner_dataset["X19"].unique()))
```
