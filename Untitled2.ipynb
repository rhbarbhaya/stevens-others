```python
import pandas as pd
url = 'https://s3.amazonaws.com/orim-misc-data/assessment/subscribers.csv'
dataset = pd.read_csv(url)
dataset["Signup"] = pd.to_datetime(dataset["Signup"])
```

```python
filtered_df = pd.DataFrame()
filtered_df["Signup_year"] = dataset["Signup"].dt.year
filtered_df["Signup_month"] = dataset["Signup"].dt.month
filtered_df["referral_source"] = dataset["referral_source"]
filtered_df
```

```python
df = filtered_df.assign(sortkey = filtered_df.groupby(["referral_source", "Signup_year", "Signup_month"])["Signup_month"].transform("count").sort_values(["sortkey", "referral_source", "Signup_year", "Signup_month"]), ascending = [False, True, False, True]).set_index(["referral_source", "Signup_year", "Signup_month"]).drop("sortkey", axis = 1)
df
```
